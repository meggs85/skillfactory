# Предобработка

```python
def CountComplex(field):
    global data
    print("CountComplex for", field)
    cnt = len(data)
    r = range(0, cnt)
    unique_dict = {}
    values_serie = data[field].apply(lambda x: x.split('|'))
    for i in r:
        row = values_serie.iloc[i]
        if i%100 == 0:
            print(i)
        for x in row:
            data.loc[i, x] = True
            if x in unique_dict.keys():
                unique_dict[x] += 1
            else:
                unique_dict[x] = 1
    print("done")
    return unique_dict


answers = {} # создадим словарь для ответов


data['profit'] = data['revenue']-data['budget']



genres_dict = CountComplex('genres')
cast_dict = CountComplex('cast')
production_companies_dict = CountComplex('production_companies')
#print(genres_dict)
```

# 15. Фильмы с каким актером принесли самые высокие кассовые сборы в 2012 году? 

```python
data = data[data.release_year==2012]

res = pd.DataFrame(cast_dict.items(), columns=['actor', 'revenue'])

for actor in cast_dict.keys():
    try:
        rev = data.groupby([actor])['revenue'].sum().iloc[0]
        res.loc[res.actor==actor, 'revenue'] = rev
    except IndexError:
        res.loc[res.actor==actor, 'revenue'] = 0
display(res.sort_values('revenue'))
    
answers['15'] = 'Joss Whedon'
```

# 16. Какой актер снялся в большем количестве высокобюджетных фильмов?

```python
#Примечание: в фильмах, где бюджет выше среднего по данной выборке.
df = data[data.budget > data['budget'].mean()]
actors = df['cast'].apply(lambda x: x.split('|'))
actors_list = {}
for xx in actors:
    for x in xx:
        try:
            actors_list[x] += 1
        except KeyError:
            actors_list[x] = 1

d = pd.DataFrame(actors_list.items(), columns=['actor', 'num']).sort_values('num')
print(d)
answers['16'] = 'Matt Damon'
```

# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
data['namelen'] = (data['original_title'].apply(lambda x: len(x))).astype(int)

res = pd.DataFrame(production_companies_dict.items(), columns=['comp', 'num'])
display(res)
for x in production_companies_dict.items():
    comp = x[0]
    mean = data[data[comp]==True]['namelen'].mean()
    res.loc[res.comp==comp, 'meanlen'] = mean

display(res.sort_values(by=['meanlen']))
answers['24'] = 'Four By Two Productions'
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
import itertools
combs = list(itertools.combinations(cast_dict, 2))
df_combs = pd.DataFrame(combs)
df_combs['freq'] = pd.Series([0 for x in range(len(df_combs.index))])
display(df_combs)
i = 0
for pair in combs:
    one = pair[0]
    two = pair[1]
    films = data[(data[one]==True) & (data[two]==True)]
    n = len(films)
    df_combs.loc[(df_combs[0]==one) & ((df_combs[1]==two)),'freq']=n
    i += 1
    if i % 1000 == 0:
        print(i)
display(df_combs.sort_values('freq'))

```

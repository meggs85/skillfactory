```python
def safe_sum(a,b):
    try:
        return a+b
    except TypeError:
        print("Can't sum {} and {}".format(a,b))
        return 0
         
safe_sum(1, 2)
safe_sum(5, 'a')
```

```python
from copy import copy

def remove_dups(values):
    rv = []
    for i in range(0,len(values)):
        if values[i] in rv:
            pass
        else:
            rv += [values[i]]
        print(i, ':', rv)
    return rv

remove_dups([1, 12, 4, 1, 4, 8])
```

```python
import requests  
response = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')  
#print(response.text)
currencies = response.json()
#print()
print(currencies['Valute']['CZK']['Name'])
```

```python
def currency_name(currency_id):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']
    for data in response.values():
        if data['ID']==currency_id:            
            return data['Name']
    return None

currency_name('R01700J')
```

```python
from bs4 import BeautifulSoup    
import requests    


def wiki_header(url):
    response = requests.get(url)    
    return BeautifulSoup(response.text, 'html.parser').find('h1').text

wiki_header('https://en.wikipedia.org/wiki/Operating_system')
```

```python
def get_actors(url):
    response = requests.get(url)    
    links = BeautifulSoup(response.text, 'html.parser').find('ul', class_='styles_list__I97eu').find_all('a')
    return [x.text for x in links]

get_actors('https://www.kinopoisk.ru/film/195334/')
```
